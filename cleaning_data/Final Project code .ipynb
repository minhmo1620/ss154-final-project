# Cleanning the data

```python
import pandas as pd
```

## Data 1

```python
# Read csv file
# Original data: https://www.kaggle.com/wsj/college-salaries?select=salaries-by-region.csv
data1 = pd.read_csv('salaries-by-region.csv', error_bad_lines=False)
```

```python
# Overview of the data
data1.head(20)
```

```python
# Rename columns
data1 = data1.rename(
    columns = {
        "School Name": 'name', 
        'Starting Median Salary':'starting_median_salary'
    }
)
```

```python
# Reformatting starting_median_salary from '$AAAA'(str) to AAA.00(float)
for i in range (len(data1)):
    row = data1['starting_median_salary'][i][1:]
    lst = list(row.split(','))
    data1['starting_median_salary'][i] = float(''.join(lst))
```

```python
# Necessary columns for analyze
kept_cols = [
    'name',
    'starting_median_salary',
]

# Drop unnecessary columns
for col in list(data1.columns):
    if col not in kept_cols:
        data1.drop(col, axis='columns', inplace=True)
```

## Data 2: College characteristics data

```python
# Load data 2 with college demographics
# Orginal data link: https://opportunityinsights.org/data/
# Name data: College Level Characteristics from the IPEDS Database and the College Scorecard
data2 = pd.read_csv('mrc_table10.csv')
```

```python
data2.head(10)
```

```python
# Show all columns of data2
data2.columns
```

```python
# Necessary columns for analyze
kept_cols = [
    'name',
    'starting_median_salary',
    'zip',
    'tier',
    'sat_avg_2013',
    'asian_or_pacific_share_fall_2000',
    'black_share_fall_2000',
    'hisp_share_fall_2000',
    'alien_share_fall_2000'
]

# Drop unnecessary columns
for col in list(data2.columns):
    if col not in kept_cols:
        data2.drop(col, axis='columns', inplace=True)
```

```python
# Rename columns
data2 = data2.rename(
    columns = {
        'asian_or_pacific_share_fall_2000': 'asian_or_pacific_share',
        'black_share_fall_2000':'black_share',
        'hisp_share_fall_2000': 'hisp_share',
        'alien_share_fall_2000': 'alien_share'
    }
)
```

## Data 3: Collge by city and population

```python
# Import data
url = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTcLn5ahGpzQAAK4MZqlZ-vcsR2OjjxECPN4hOMnegZphhmKP6VGG4GhXVBV9qnjU4azuRC_OaURp8X/pub?gid=1955362680&single=true&output=csv'
data3 = pd.read_csv(url)
```

```python
data3.head(10)
```

```python
# Formatting column population_in_2010
# Changing the string to float
for i in range (len(data3)):
    row = data3['population_in_2010'][i]
    lst = list(row.split(','))
    data3['population_in_2010'][i] = int(''.join(lst))
```

## Data 4: Parent income by college

```python
# Load data 2 with parent income
# Orginal data link: https://opportunityinsights.org/data/
# Name of dataset: Baseline Cross-Sectional Estimates of Child and Parent Income Distributions by College
data4 = pd.read_csv('mrc_table2.csv')
```

```python
data4.columns
```

```python
# Necessary columns for analyze
kept_cols = [
    'name',
    'par_median'
]

# Drop unnecessary columns
for col in list(data4.columns):
    if col not in kept_cols:
        data4.drop(col, axis='columns', inplace=True)
```

```python
data4
```

## Merge datasets

```python
## Merge data1 and data2 by college name, only keeps colleges in both dataset
data5 = pd.merge(data1, data2, on='name', how='inner')
```

```python
## Merge data3 and data4 by college name
data6 = pd.merge(data3, data4, on='name', how='inner')
```

```python
## Merge data5 and data6 by college name
data = pd.merge(data5, data6, on='name', how='inner')
```

```python
## Drop NA row
data = data.dropna(axis = 0, how ='any')
```
